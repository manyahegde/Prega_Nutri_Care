```python
import requests
from bs4 import BeautifulSoup
import json

# URL of the webpage to scrape
url = "https://www.acog.org/womens-health/faqs/pregnancy-with-type-1-or-type-2-diabetes"

# Send an HTTP request to the URL
response = requests.get(url)

# Parse the HTML content of the webpage
soup = BeautifulSoup(response.text, "html.parser")

soup
```

```python
# Find the FAQs section
faq_section = soup.find_all("button", class_="accordion-trigger")
```

```python
faq_section
```

```python

faq_ans = soup.find_all("div", class_="accordion-content wysiwyg-content")
```

```python
faq_ans
```

```python
type(faq_section)
```

```python
import json

# Assuming faq_qstn and faq_ans are bs4.element.ResultSet
# Convert them to lists to access elements by index
faq_qstn_list = list(faq_section)
faq_ans_list = list(faq_ans)

# Initialize an empty list to store question-answer pairs
faq_list = []

# Iterate through the pairs of questions and answers
for qstn, ans in zip(faq_qstn_list, faq_ans_list):
    # Extract the text of the question and answer
    qstn_text = qstn.get_text()
    ans_text = ans.get_text()
    
    # Extract the tag of the answer
    ans_tag = ans.name

    # Create a dictionary representing the question-answer pair
    faq_pair = {
        "question": qstn_text,
        "answer": ans_text,
        "answer_tag": ans_tag
    }
    
    # Append the pair to the list
    faq_list.append(faq_pair)

# Write the list of question-answer pairs to a JSON file
with open("faq_data.json", "w") as json_file:
    json.dump(faq_list, json_file, indent=4)
```
